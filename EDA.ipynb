```python
genres = ['Action', 'Adventure', 'Animation &amp; Modeling',
       'Audio Production', 'Casual', 'Design &amp; Illustration',
       'Early Access', 'Education', 'Free to Play', 'Indie',
       'Massively Multiplayer', 'Photo Editing', 'RPG', 'Racing', 'Simulation',
       'Software Training', 'Sports', 'Strategy', 'Utilities',
       'Video Production', 'Web Publishing']


relevant_columns = ['year', 'playtime_forever'] + genres

# Filtra el DataFrame según el rango de años deseado
df_filtered = df[(df['year'] >= 2001) & (df['year'] <= 2013)][relevant_columns].copy()

# Agrupa por año y suma las horas jugadas por género
df_grouped = df_filtered.groupby('year')[genres].sum()
df_grouped['total_playtime'] = df_grouped[genres].sum(axis=1)
df_grouped = df_grouped.sort_values('total_playtime', ascending=False)

# Define una paleta de colores personalizada
custom_palette = sns.color_palette("husl", len(genres))

# Visualiza un gráfico de barras apiladas con la paleta de colores personalizada
df_grouped[genres].plot(kind='bar', stacked=True, figsize=(12, 6), color=custom_palette)
plt.xlabel('Año')
plt.ylabel('Horas Jugadas')
plt.title('Horas Jugadas por Año y Género')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

RANKING DE JUEGIS Y HORAS JUGADAS

```python
# Verificamos el Ranking de los juegos

# Agrupa por nombre del juego y suma las horas jugadas
df_games_grouped = df.groupby('item_name')['playtime_forever'].sum()

# Ordena los juegos por horas jugadas en orden descendente
df_games_ranked = df_games_grouped.sort_values(ascending=False)

# Muestra los primeros elementos del ranking (por ejemplo, los primeros 10 juegos)
print(df_games_ranked.head(10))
```

```python
# Define la cantidad de juegos que quieres mostrar en el gráfico
num_top_games = 10

# Selecciona los juegos principales según el ranking
top_games = df_games_ranked.head(num_top_games)

# Crea una paleta de colores personalizada
custom_palette = sns.color_palette("husl", num_top_games)

# Crea un gráfico de barras para visualizar los juegos principales y sus horas jugadas
plt.figure(figsize=(10, 6))
top_games.plot(kind='bar', color=custom_palette)
plt.title('Ranking de Juegos por Horas Jugadas')
plt.xlabel('Juego')
plt.ylabel('Horas Jugadas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

CALIFICACION PROMEDIO REVIEWS


```python
# Agrupa por nombre del juego y calcula la media de las calificaciones de las reviews
df_reviews_grouped = df.groupby('item_name')['review'].mean()

# Ordena los juegos por calificación promedio en orden descendente
df_reviews_ranked = df_reviews_grouped.sort_values(ascending=False)

# Muestra los primeros elementos del ranking (por ejemplo, los primeros 10 juegos)
print(df_reviews_ranked.head(10))
```

```python
# Filtra los juegos que tienen calificaciones de reviews válidas (no nulas)
df_valid_reviews = df.dropna(subset=['review'])

# Calcula la calificación promedio de las reviews por juego
df_reviews_avg = df_valid_reviews.groupby('item_name')['review'].mean().sort_values(ascending=False)

# Obtén los primeros 10 juegos con las calificaciones promedio más altas
top_10_reviews_avg = df_reviews_avg.head(10)

# Configura la paleta de colores
colors = sns.color_palette("husl", len(top_10_reviews_avg))

# Crea el gráfico de barras horizontal
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_reviews_avg.values, y=top_10_reviews_avg.index, palette=colors)
plt.title('Top 10 Juegos por Calificación Promedio de Reviews')
plt.xlabel('Calificación Promedio')
plt.ylabel('Juego')
plt.grid(True, axis='x')
plt.show()
```

RECUENTO DE REVIEWS POR AÑO 

```python
# Agrupa por año y cuenta el número de reviews por año
reviews_count_by_year = df.groupby('posted_year').size()

# Configura la paleta de colores
colors = sns.color_palette("husl", len(reviews_count_by_year))

# Crea el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=reviews_count_by_year.index, y=reviews_count_by_year.values, palette=colors)
plt.title('Recuento de Reviews por Año')
plt.xlabel('Año')
plt.ylabel('Recuento de Reviews')
plt.grid(True, axis='y')
plt.xticks(rotation=45)
plt.show()
```

JUEGOS MENOS RECOMENDADOS 

```python
# Filtra los juegos con recomendaciones negativas
df_less_recommended_games = df[df['recommend'] == False]

# Agrupa por nombre del juego y cuenta el número de veces que fue menos recomendado
less_recommended_games_count = df_less_recommended_games.groupby('item_name').size().sort_values()

# Toma los 10 juegos con menos recomendaciones
top_10_less_recommended_games = less_recommended_games_count.head(10)

# Configura una paleta de colores
colors = sns.color_palette("husl", len(top_10_less_recommended_games))

# Crea el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_less_recommended_games.values, y=top_10_less_recommended_games.index, palette=colors)
plt.title('Top 10 Juegos Menos Recomendados')
plt.xlabel('Número de Veces Menos Recomendado')
plt.ylabel('Juego')
plt.grid(True, axis='x')
plt.show()
```

EMPRESA DESARROLLADORA MENOS RECOMENDADO

```python
# Filtra los desarrolladores con recomendaciones negativas
df_less_recommended_developers = df[df['recommend'] == False]

# Agrupa por nombre del desarrollador y cuenta el número de veces que fue menos recomendado
less_recommended_developers_count = df_less_recommended_developers.groupby('developer').size().sort_values()

# Toma los 10 desarrolladores con menos recomendaciones
top_10_less_recommended_developers = less_recommended_developers_count.head(10)

# Crea el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_less_recommended_developers.values, y=top_10_less_recommended_developers.index, palette=colors)
plt.title('Empresa Desarrolladora Menos Recomendados')
plt.xlabel('Número de Veces Menos Recomendado')
plt.ylabel('Desarrollador')
plt.grid(True, axis='x')
plt.show()
```
